1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()



```python
def myreduce(fun, l, init=None):
    it = iter(l)
    
    if init == None:
        value = next(it)
    else:
        value = init

    for i in it:
        value = fun(value, i)

    return value

l = [1, 2, 3, 46, 5, 33, 99]
print(myreduce(lambda x, y : x+y, l))
print(myreduce(lambda x, y : x+y, l, 5))
```

1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()



```python
def myfilter(fun, l):
    for i in l:
        if fun(i):
             yield i

for i in iter(myfilter(lambda x : x%5 == 0, l)):
    print(i)
```

3. Implement a function longestWord() that takes a list of words and returns the longest one.



```python
def longestWord(l):
    n = [len(i) for i in l]
    m = n[0]
    for i in range(1,len(n)):
        if n[i] > m:
            m = n[i]
    return l[n.index(m)]

l = ['abcd', 'abcdekjsnfkjsdsdjff', 'ooobhghsg', 'hahsdjggsah']

print(longestWord(l))
```

1.1 Write a Python Program(with class concepts) to find the area of the triangle.



```python
class shape:
    def getSides(self):
        self.a = int(input('Enter length of 1st side: '))
        self.b = int(input('Enter length of 2nd side: '))
        self.c = int(input('Enter length of 3rd side: '))

class triangle(shape):
    def Area(self):
        self.getSides()
        s = (self.a + self.b + self.c) / 2
        self.area = (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5
        print(f'area = {self.area}')
        
t1 = triangle()
t1.Area()
```

1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.



```python
def filter_long_words(l, n):
    return [i for i in l if len(i) > n]

print(filter_long_words(l, 4))
```

2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.



```python
def isvowel(c):
    if c in ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']:
        return True
    else:
        return False
    
print(isvowel('a'))
print(isvowel('p'))
```
